```python
import pandas as pd
import numpy as np
import matplotlib as plt
import warnings
from pylab import rcParams
import seaborn as sns
from scipy import stats
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
from matplotlib import pyplot as plt
```

```python
# formats plots to view along with
%matplotlib inline
np.random.seed(8)
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 8,8
rcParams['font.size'] = 12
sns.set
```

```python
# I made a function that uses t-test from scipy.stats module
# to preform t-tests on entered data and prints the result
def compare_2_groups(arr_1, arr_2,alpha,sample_size):
    stat, p = ttest_ind(arr_1,arr_2)
    print(('Statistics=%.3f, p = %.3f' % (stat, p)))
    if p > alpha:
        print('Same distribution (fail to reject H0)')
    else:
        print('Difference distribution (reject H0)')
```

```python
# I used the same_day_algo fucntion to create csv files
# of aligned data for each index I want to test
# I will first test the S&P500 volatility deviations
sandp = pd.read_csv('same day sandp.csv')
sample_size = len(sandp)
test_volatility = sandp['Wills Volatility Measure']
controlled_volatility = sandp['Mean volatility on this day other years']
compare_2_groups(test_volatility, controlled_volatility, 0.05, sample_size)
```

```python
sns.distplot(test_volatility)
sns.distplot(controlled_volatility)
plt.savefig('sandp_ttest.png')
```

```python
# S&P500 Rejects H0
```
